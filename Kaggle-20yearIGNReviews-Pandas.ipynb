```python
#applying real world filters
score_filter = reviews["score"] > 7
score_filter
```

```python
#another filter
xbox_one_filter = (reviews["score"] > 7) & (reviews["platform"] == "Xbox One")
filtered_reviews = reviews[xbox_one_filter]
filtered_reviews.head()
```
